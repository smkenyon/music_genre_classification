```python
names = data['names'].to_list()
x = []
for name in names:
    s = name.split("\\")
    x.append(int(s[-1][:-4]))  # trim .mp3 and get the id
u = pd.Series(x, name='ids')
print(type(u))
print(type(data))
data = pd.concat([data, u], axis=1)
```

## Load up the tracks metadata csv file and genres csv file to act as lookup tables ##

```python
metadata_dir = r'D:\Analytics\fma_metadata'
tracks_metadata = r'D:/Analytics/fma_metadata/raw_tracks.csv'
tmdf = pd.read_csv(tracks_metadata)
```

```python
id_and_genre = tmdf[['track_id','track_genres']]
genres = id_and_genre['track_genres'].to_list()
print(genres[0])
print(type(genres[0]))
```

```python
x = (eval(genres[0])[0])
x['genre_title']
```

```python
y = []
for genre in genres:
    if genre is not np.nan:
        s = eval(genre)[0]
        y.append(s['genre_title'])
    else:
        y.append('missing')
```

```python
z = pd.Series(y, name='genre')
```

```python
id_and_genre = pd.concat([id_and_genre, z], axis=1)
```

```python
id_and_genre
```

```python
blah = data.join(id_and_genre.set_index('track_id'), on='ids')
blah
```

```python
blah.to_csv('fma_small_with_genres.csv')
```

## Collapsing Genres ##
Our next step is to collapse the genres to a smaller space.

The following are our genres: 

```python
gs = blah['genre']
unique_genres = []
for genre in gs.to_list():
    if genre not in unique_genres:
        unique_genres.append(genre)
genre_series = pd.Series(unique_genres)
genre_series.to_csv('genres_lut.csv')
```

```python
genre_lut = pd.read_csv('genres_lut.csv')
```

```python
genre_lut
```

```python
## Now that we have the lookup table, we need to append to the dataframe the new genre, by looking up the old genre in the LUT, and then appending the new genre
```

```python
lut = pd.Series(genre_lut['new genre'].values, index=genre_lut['old genre']).to_dict()
print(type(lut))
```

```python
g = blah['genre'].to_list()
```

```python
flattened_genre = []
for genre in g:
    flattened_genre.append(lut[genre])
blah = pd.concat([blah, pd.Series(flattened_genre, name='flattened genre')], axis=1)
blah
```

```python
blah.to_csv('fma_small_with_genres_flattend.csv')
```
