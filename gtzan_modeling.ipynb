## Importing the GTZAN Dataset and Data Cleaning ##
We'll begin by importing the GTZAN dataset stored in a csv file into
a Pandas Dataframe. The code block below shows all of the imports that we will be using in this notebook.

```python
import numpy as np
import scipy as sp
import pandas as pd
import os
```

```python
working_dir = r'D:\Analytics'
print(os.getcwd())  # check to make sure in correct directory
gtzan_csv = 'extracted_audio_features_gtzan.csv'
gtzan_frame = pd.read_csv(gtzan_csv)
print(gtzan_frame)
```

On import, we see that we have an unnamed index column that must be removed. In addition, we should build a new attribute named "genres" that we build from the "names" column. We can extract the genre from names by performing simple string operations.

```python
names = gtzan_frame['names'].tolist()
names[0].split('\\')[1]
genres = [genre.split('\\')[1] for genre in names]
genres = pd.Series(genres, name='genres')
gtzan_frame = pd.concat([gtzan_frame, genres], axis=1)
print(gtzan_frame)
gtzan_frame.to_csv('gtzan_with_genres.csv')
```
